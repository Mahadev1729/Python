```python
## PairPlot
import seaborn as sns
import matplotlib.pyplot as plt 
tips=sns.load_dataset('tips')
tips
```

```python
# pairplot
sns.pairplot(tips)
```

```python
## heatmap
corr=tips[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
