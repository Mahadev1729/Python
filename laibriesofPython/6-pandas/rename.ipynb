```python
import pandas as pd 
df=pd.read_csv('data_result.csv')
df=df.rename(columns={'total ':'Total Marks'})
df
```

```python
## Change datatype
df['Total']=df['Total Marks'].astype(float)
df.head()
df['Total'].sum
```

```python
## apply function
df['Total_grace']=df['Total'].apply(lambda x:x+1)
df
```

```python
df['increase_marks']=df['Total'].apply(lambda x:x*2)
df
```

```python
## data aggregating and grouping
grouped_mean=df.groupby('Total')
grouped_mean
```

```python
sorted_val=df.sort_values(by=['Total'])

sorted_val.head(150)
```
