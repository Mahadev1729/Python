```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]

##create a line plot
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('basic')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,9,10]
y=[1,4,9,16,25,30]

## custmized line plot
plt.plot(x,y,color='red',linestyle='-',marker='o',linewidth=3)
plt.grid(True)
```
