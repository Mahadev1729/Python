```python
import matplotlib.pyplot as plt # type: ignore
## Bar Plots
categories=['A','B','C','D','E']
values=[5,7,3,8,6]
# create Barr plot
plt.bar(categories,values,color='purple')
plt.title('Bar Plot')
plt.xlabel('categories')
plt.ylabel('values')
```
