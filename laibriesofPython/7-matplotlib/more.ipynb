```python
##Multiple Plots
## Sample Data
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y1=np.array([1,4,9,16,25])
y2=np.array([1,2,3,4,5])

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.grid(True)
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.grid(True)
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(y2,x,color='blue')
plt.grid(True)
plt.title('plot 3')

plt.show()
```
